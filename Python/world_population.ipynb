## Exploratory Data Analysis

#### Importing libraries

```python
import pandas as pd
import matplotlib as plt
```

#### Importing .csv file

```python
df=pd.read_csv('world_population.csv')
df
```

#### Formating numbers to float, to two decimals

```python
pd.set_option('display.float_format', lambda x:'%.2f' % x)
```

#### Checking table info

```python
df.info()
```

#### Overview of data

```python
df.describe()
```

#### Counting NULL values per column

```python
df.isnull().sum()
```

#### Counting unique values per column

```python
df.nunique()
```

#### Filtering only columns with object type of data 

```python
df.select_dtypes(include='object')
```

#### Showing Top10 countries by 2022 population

```python
df.sort_values(by='2022 Population', ascending=False).head(10)
```

#### Searching for correlations between columns

```python
df.corr()
```

#### Grouping countries by the continent, counting average values and sorting them in descending order by 2022 population

```python
df.groupby('Continent').mean().sort_values(by='2022 Population', ascending=False)
```

#### Showing countries of Oceania

```python
df[df['Continent'].str.contains('Oceania')]
```

#### Showing names of columns

```python
df.columns
```

#### Visualization of population change data by continets

```python
df2=df.groupby('Continent')['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population'].mean().sort_values(by='2022 Population', ascending=False)
df2=df2.transpose()
df2.plot(figsize=(20,10))
```

```python
df.boxplot(figsize=(20,10))
```
