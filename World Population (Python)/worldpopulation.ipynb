## Exploratory Data Analysis

#### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
plt.style.use('ggplot')
```

#### Importing .csv file

```python
df=pd.read_csv('/kaggle/input/projects/World Population (Python)/world_population.csv')
df
```

#### Formating numbers to float, to two decimals

```python
pd.set_option('display.float_format', lambda x:'%.2f' % x)
```

#### Checking table info

```python
df.info()
```

#### Performing statistical calculations

```python
df.describe()
```

#### Counting NULL values per column

```python
df.isnull().sum()
```

#### Counting unique values per column

```python
df.nunique()
```

#### Filtering only columns with object type of data 

```python
df.select_dtypes(include='object')
```

#### Showing Top20 countries by 2022 population

```python
top20 = df[['Country','2022 Population']].sort_values(by= '2022 Population', ascending=False) \
.head(20)
```

```python
top20.sort_values(by= '2022 Population', ascending=True) \
.plot(kind='barh', x='Country', title='Top 20 Countries by Population in 2022', xlabel='Number of People in billion', legend=None)
plt.show()
```

#### Searching for correlations between columns

```python
df_corr=df.corr(numeric_only=True)
sns.heatmap(df_corr, annot=True, fmt=".1g", linewidths=1)
```

#### Grouping countries by the continent, counting average values and sorting them in descending order by 2022 population

```python
df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```

#### Showing countries of Oceania

```python
df[df['Continent'].str.contains('Oceania')]
```

#### Visualization of population change data by continets

```python
df2=df.groupby('Continent')['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population'].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
df2=df2.transpose()
df2
df2.plot(figsize=(20,10), title='Average population by Continent from 1970 to 2022')
```
